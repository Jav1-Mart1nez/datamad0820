#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
response = requests.get("https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers")
html = response.text
soup = BeautifulSoup(html, "html.parser")
top_lang = soup.find_all("table")[2]
top_lang
````

````python
def procesaFila(fila):
    celdas = fila.find_all("td")
    return {
        "language":celdas[1].find("a").text,
        "native speakers":celdas[2].text,
    }

fila = top_lang.find_all("tr")[1]
print(fila)
procesaFila(fila)
````

````python
data = [procesaFila(tag) for tag in top_lang.find_all("tr")[1:11]]
print(data)
df = pd.DataFrame(data)
df
````
